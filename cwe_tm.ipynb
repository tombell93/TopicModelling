```python
from bertopic import BERTopic

import pandas as pd
df = pd.read_csv("data/1000_cwes.csv")
df
```

```python
docs = df.Description.tolist()
docs[0]
```

```python
topic_model = BERTopic(embedding_model="all-MiniLM-L6-v2")
topics, probs = topic_model.fit_transform(docs)
```

```python
topic_model.get_topic_info()
```

```python
topic_model.get_topic(0)
```

```python
topic_model.get_representative_docs(1)
```

```python
df = pd.DataFrame({"topic": topics, "document": docs})
df
```

```python
fig = topic_model.visualize_topics()
fig.write_html('outputs/cwe/cwe_topics.html', auto_open=False)
```

```python
fig = topic_model.visualize_barchart()
fig.write_html('outputs/cwe/cwe_barchart.html', auto_open=False)
```

```python
hierarchical_topics = topic_model.hierarchical_topics(docs)
fig = topic_model.visualize_hierarchy(hierarchical_topics=hierarchical_topics)
fig.write_html('outputs/cwe/cwe_hierarchy.html', auto_open=False)
```
